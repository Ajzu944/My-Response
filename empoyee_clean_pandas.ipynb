```python
for i in ["FirstName", "LastName", "Department", "City"]:
    df[i]=df[i].str.title().str.strip()
    
df["Email"]=df["Email"].str.lower()
df["Phone"]=df["Phone"].str.replace(r"\D", "", regex=True)
df["Phone"]=df["Phone"].str.replace(r"^1", "", regex=True)
df["Phone"] = df["Phone"].apply(lambda x: x if len(x) == 10 else "None")
df["Phone"]= df["Phone"].str.replace(r"(\d{3})(\d{3})(\d{4})", r"\1-\2-\3",regex=True)

df
```

```python
df.to_csv("employees_cleaned.csv")
print("Cleaned and saved succesfully.")
```
